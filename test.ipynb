```python
import pandas as pd

file_path = "/Users/benwong/Documents/y4s2/dsa4262/dsa4262-group-3/HBSC_data/HBSC2018OAed1.1.csv"

df = pd.read_csv(
    file_path,
    sep=None,
    engine="python"
)

print(df.shape)
print(df.head())
print(df.columns.tolist())
```

```python
overview = pd.DataFrame({
    "column": df.columns,
    "dtype": df.dtypes.astype(str),
    "non_null": df.notna().sum().values,
    "missing": df.isna().sum().values,
    "missing_pct": (df.isna().mean() * 100).round(2).values,
    "n_unique": df.nunique(dropna=True).values
})

print("Shape:", df.shape)
display(overview.sort_values(["n_unique", "missing_pct"]))
```
